```python
expenses=['monthly_rent','school_fees','college_fees','travel_expenses','groceries_utilities','other_monthly_expenses']
```

```python
avg_expennses=df[expenses].mean()
```

```python
avg_expennses
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=avg_expennses.index,y=avg_expennses.values)
plt.xlabel("Expense Type")
plt.ylabel("Average Monthly Expense")
plt.title("Average Monthly Expenses")
plt.grid()
plt.show()
```

```python
percent=(avg_expennses/avg_expennses.sum())*100
plt.figure(figsize=(8,8))
plt.pie(percent, labels=percent.index, autopct='%1.1f%%', startangle=140)
plt.title("Percentage Distribution of Monthly Expenses")
plt.show()
```

```python
expense_ratio=(df['total_expenses']/df['monthly_salary'])*100
expense_ratio
```

```python
plt.figure(figsize=(10,6))
sns.histplot(expense_ratio, bins=30, kde=True)
plt.xlabel("Expense to Salary Ratio (%)")
plt.title("Distribution of Expense to Salary Ratio")
# plt.grid()
plt.show()
```

```python
df['expense_to_income_ratio']=np.where(df['monthly_salary']>0,df['total_expenses']/df['monthly_salary'],np.nan)
```

## Employment and Financial Stability

How does EMI eligibility vary across employment types?

```python
sns.countplot(x='employment_type', hue='emi_eligibility', data=df)
plt.title("Employment Type vs EMI Eligibility")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.show()
```

Is there a relationship between years of employment and credit score

```python
sns.scatterplot(x='years_of_employment', y='credit_score', hue='emi_eligibility', data=df)
plt.title("Years of Employment vs Credit Score by EMI Eligibility")
plt.xlabel("Years of Employment")
plt.ylabel("Credit Score")
plt.show() 
```

```python
df
```

```python
high_credit=df[df['credit_score']>850]
```

```python
high_credit
```

```python
df = df[(df['credit_score'] >= 300) & (df['credit_score'] <= 850)]
```

```python
df[df['credit_score']>850]
```

```python
df.info()
```

```python
sns.scatterplot(x='years_of_employment', y='credit_score', hue='emi_eligibility', data=df)
plt.title("Years of Employment vs Credit Score by EMI Eligibility")
plt.xlabel("Years of Employment")
plt.ylabel("Credit Score")
plt.show() 
```

```python
bins=[0,2,5,10,20,df['years_of_employment'].max()]
labels = ['0–2 yrs', '3–5 yrs', '6–10 yrs', '11–20 yrs', '20+ yrs']
df['employment_group'] = pd.cut(df['years_of_employment'], bins=bins, labels=labels, include_lowest=True)
# df['employment_group']=pd.cut(df['years_of_employment'], bins=bins, labels=labels, right=False)
emi_tenure_group=df.groupby('employment_group')['max_monthly_emi'].mean().reset_index()
emi_tenure_group
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=emi_tenure_group['employment_group'], y=emi_tenure_group['max_monthly_emi'], palette='viridis')
plt.title("Max Monthly EMI by Employment Group")
plt.xlabel("Employment Group")
plt.ylabel("Max Monthly EMI")
plt.show()
```
